7. 7 ***(Challenge Project: Reimplement NumPy array Output)*** You saw that NumPy outputs two-dimensional arrays in a nice column-based format that right-aligns every element in a field width. The field width's size is determined by the array element value that requires the most character positions to display. To understand how powerful it is to have this formatting simply built-in, write a function that reimplements NumPy's array formatting for two-dimensional arrays using loops. Assume the array contains only positive integer values.

````python
import numpy as np

test = np.arange(1,31).reshape(5,6)
test_two = np.random.randint(1,319,30).reshape(5,6)


def align_(sequence: np.array) -> print:
    """
        displays the array in tabular format by right-aligning its values 
    Args:
        sequence (Sequence): N-dimensional array (ndarray)
        
    Returns: None 
    """

    length = str(sequence.max()).__len__()
    
    for array in sequence:
        print(*(f"{element:>{length}}" for element in array))
        
align_(test)
print()


align_(test_two)

````

7. 8 ***(Challenge Project: Reimplement DataFrame Output)*** You saw that pandas displays DataFrames in an attractive column-based format with row and column labels. The values within each column are right aligned in the same field width, which is determined by that column’s widest value. To understand how powerful it is to have this formatting built-in, write a function that reimplements DataFrame formatting using loops. Assume the DataFrame contains only positive integer values and that both the row and column labels are each integer values beginning at 0.

````python

import numpy as np
import pandas as pd

test = pd.DataFrame(np.arange(1,31).reshape(5,6))
test_two = pd.DataFrame(np.random.randint(1,319,30).reshape(5,6))
test_three = pd.DataFrame(np.random.randint(1,2999, 30).reshape(5,6), 
                          columns= map(chr,range(65,71)), index= map(chr,range(71,76)))

def align_DF(Data:...)  -> None:
    """ 
        displays the data frame with its indices aligned to the left\\
        and its columns and values aligned to the right.

    Args:
        Data (DataFrame): DataFrame[integers]
    """
    
    length = str(Data.values.max()).__len__()
    
    print(f"{'':<{length}}",*(f"{column:>{length}} " for column in Data.columns))
   
    for index, array in zip(Data.index, Data.values):
        
        print(f"{index:<{length}}", *(f"{values:>{length}} " for values in array))
    
align_DF(test)
print()
align_DF(test_two)
print()
align_DF(test_three)
````

7. 16 ***(Research: NumPy tile Function)*** Research NumPy's tile function. Use the tile, vstack, and hstack functions to create a checkerboard pattern. Each “square” should consist of 4 zeros or 4 ones.

````python
import numpy as np

zeros = np.zeros(4, dtype= np.int16).reshape(2,2)
ones = np.ones(4, dtype= np.int16).reshape(2,2)

squares_v = np.hstack((ones,zeros))
squares_h = np.hstack((zeros,ones))

both_squares  = np.vstack((squares_h,squares_v))

checkerboard = np.tile(both_squares ,(4,4))

checkerboard
````

````python


from numpy import arange, random


def median(array, axis = 1):
    ''''It takes an array, sorts it, finds the middle value, and returns the middle value if the array has
    an odd number of elements, or the average of the two middle values if the array has an even number
    of elements

    Parameters
    ----------
    array
        The array to find the median of.
    axis, optional
        The axis along which the median is to be computed
        0 or 1 (default), depending on whether you want to find the mode of each column or row.
    
    Returns
    -------
        The median of the array.
    
    '''
    import numpy as np

    array = np.sort(array) if axis else np.sort(array.T)
    
    # Finding the number of columns in the array.
    middle = (array.shape[len(array.shape) - 1]
              ) if array.ndim > 1 else (array.size)
    
    dim = array.ndim - 1 

    if (middle & 1): # if the number of column is odd
        return array[..., middle // 2]
    
    else:
        middle //= 2
        return array[..., [middle, middle - 1]].sum(axis=dim) / 2


def mode(array, axis = None):
    ''' returns the mode of the array along the specified axis

    Parameters
    ----------
    array
        The array to find the mode of.
    axis
        None(default) returns the mode of the entire array
        0 or 1, depending on whether you want to find the mode of each column or row.


    Returns
    -------
        The mode of the array.

    '''
    import numpy as np

    def extract_mode(matrix):
        '''returns the unique values in each row that have the highest frequency
        
        Parameters
        ----------
        matrix
            a 2D numpy array

        '''

        # Mapping the function np.unique to each row of the matrix, and returning the unique values
        # and their frequencies.
        uniques = list(map(lambda arr: np.unique( arr, return_counts=True), matrix))

        # return boolean values to index values more frequently in the array
        condition = lambda arrs :  (arr[1] == arr[1].max() for arr in arrs)

        # Returning the unique values in each row that have the highest frequency.
        return map(np.extract, condition(uniques), uniques)

    if axis and (array.ndim > 1): # for each row

        return np.array(list(extract_mode(array)), dtype=object)

    elif (axis == 0) and (array.ndim > 1): # for each column
        
        return np.array( list( extract_mode( array.T)), dtype=object)

    else: # mode of the entire array 
        uniques = np.unique(array, return_counts=True) 
        condition = uniques[1] == uniques[1].max()

        return np.extract(condition, uniques)


test_median_one = arange(28).reshape(4, 7)
test_median_two = arange(32).reshape(4, 2, 4)

test_mode_one = random.randint(1, 10, size=20).reshape(5, 4)
test_mode_two = random.randint(1, 10, size=36).reshape(3, 3, 4)

print(f"array test median one :\n{test_median_one}\n\nmedian function :\n{median(test_median_one)}"); print()
print(f"array test mode one :\n{test_mode_one}\n\nmode function :\n{mode(test_mode_one)}"); print()
````

````python


def make_move():
    '''The function generates a random position and verifies that it is valid, if it is valid, it moves the
    knight to that position, if it is not valid, it generates another position
    
    Returns
    -------
        a boolean value.
    
    '''
    
    from itertools import chain
    import numpy as np 
    
    def checker(current_position: np.array , new_position: np.array or tuple ) -> bool :
        '''This function verifies that the new position is valid. 
        
        Parameters
        ----------
        current_position : np.array
            the current position of the knight 
        new_position : tuple 
            the position to tester if it is valid to move the knight
        Returns
        -------
            a boolean value.
        
        '''
        
        #if the new position is either one row and two columns or two rows and one column from the current position, then it is valid 
        test = np.abs(current_position - new_position) == ([1, 2], [2, 1])  
        
        if ( (test).all(axis=1) ).any():
            
            if  chessboard[tuple(new_position)]: #verifies that the position remains valid 
                return True
            
    def accessibility(row_column: tuple or np.array ) -> bool:
        '''It checks if the knight can move to any of the 8 possible positions from the current position
        
        Parameters
        ----------
        row_column : tuple | np.array
            the current position of the knight 
        
        Returns
        -------
            A boolean value.
        
        '''
        #this narray is used to generate all possible moves. 4 when ±1 is added to the row and 4 when ±2 are added, i.e
        #[(1, 2), (-1, -2), (1, -2), (-1, 2), ], [(2, 1), (-2, -1), (2, -1),  (-2, 1)]
        possibles_ranges = np.array([[1,2],[2,1]]) #to optimize the search is tried from two positions each turn 
        counter = int()
    
        while True:
            
           # A tuple comprehension that filters the possible positions 
           #that are less than 8 and greater than 0, that is, that are not off the board and are not negative indices.
           #If one or both exists then the variable is created and it is verified that it is valid
            if (possible_positions := tuple(filter( lambda arr: arr.min() >= 0 and arr.max() < 8,
                                                   row_column + possibles_ranges))):
                
                if checker(current_position= row_column, new_position= possible_positions[0]):  
                    return False
                
                elif len(possible_positions) == 2:
                    
                    if checker(current_position= row_column, new_position= possible_positions[1]):  
                        return False    
                
            if counter < 3:
                
                if not counter or counter == 2:# #the second and fourth possible movement, i.e when count = 0 --> (row  -1, column -2) and ( row -2, column -1)
                                                #when count = 2 --> (row -1, column +2) and (row -2, column +1)
                    possibles_ranges *= -1
                    counter += 1
                       
                else:
                    possibles_ranges[:,0] *= -1 #the third possible movement, i.e (row +1, column -2) and (row +2, column -1)
                    counter += 1
            else:
                #if all valid positions have already been disabled 
                return True  #then true is returned indicating that there are no possible moves from the current position 
        
    
    while True :
        new_position = np.random.randint(8, size= 2)
        knight_Index = tuple( chain.from_iterable( (chessboard == 'K').nonzero() ))
        
        #it is verified that the new position generated is not disabled,
        if not chessboard[tuple(new_position)]: # avoiding going through the checker function to know if it is a valid movement from the current position .
            continue
        
        elif checker(current_position= knight_Index, new_position = new_position):
            
            chessboard[chessboard == 'K'] = 0
            chessboard[tuple(new_position)] = 'K'
            
        if accessibility(row_column= knight_Index): 
            return True  #if there are no more possible movements from the current position the game is finished
        
def playing(tours : int = 0, full : bool = False , BENCHMARK = False) :
    '''Play Knight's Tour.

    Parameters
    ----------
    tours : int, optional
         number of tours to play from random positions.
         
    full : bool, optional
        if True, the function will run until the knight has visited every square on the board.
    
    Returns
    -------
        if tours --> array with the number of steps taken from each square where the horse started the tours.
        
        
        If Full --> The number of moves it took to get to the end of the game and 
                        the array with the number of steps taken from each square where the horse started the tours.
    '''
    
    global chessboard
    import numpy as np
    
    
    record = np.zeros(64, dtype= np.int16) 

    while (tours if tours else full):
        
        chessboard = np.ones((8, 8), dtype= np.object_)
        position = tuple(np.random.randint(8, size= 2))
        chessboard.itemset(position, 'K')  
        
        if  make_move():
            record[chessboard.ravel() == 'K'] += (chessboard == 0).sum()
            
            if tours:
                tours -=1
                
            elif full:
                
                if (chessboard == 0).sum() == 63:
                    return record, full

                full += 1
    if not BENCHMARK:
        return record
                                        
````

##### D) PROS AND CONS OF BRUTE FORCE ALGORITHM:

##### Pros:
- It is a generic method and not limited to any specific domain of problem.
- The brute force method is deal for solving small and simpler problems.
- It is known for its simplicity and can serve as a comparison benchmark.

##### Cons:
- The brute force approach is inefficient. For real-time problems, algorithm analysis often goes above the $O(N!)$ order of growth.

- Brute force algorithms are slow.
