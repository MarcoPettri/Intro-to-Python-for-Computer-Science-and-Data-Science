8. 1 ***(Check Protection)*** Although electronic deposit has become extremely popular, payroll and accounts payable applications often print checks. A serious problem is the intentional alteration of a check amount by someone who plans to cash a check fraudulently. To prevent a dollar amount from being altered, some computerized check-writing systems employ a technique called check protection. Checks designed for printing by computer typically contain a fixed number of spaces for the printed amount. Suppose a paycheck contains eight blank spaces in which the computer is supposed to print the amount of a weekly paycheck. If the amount is large, then all eight of the spaces will be filled:
``` python
1,230.60 (check amount)
--------
01234567 (position number)

```
On the other hand, if the amount is smaller, then several of the spaces would ordinarily be left blank. For example,
``` python
  399.87
--------
01234567 

```
contains two blank spaces. If a check is printed with blank spaces, itâ€™s easier for someone to alter the amount. Check-writing systems often insert leading asterisks to prevent alteration and protect the amount as follows:
``` python
**399.87
--------
01234567 

```

Write a script that inputs a dollar amount, then prints the amount in check-protected format in a field of 10 characters with leading asterisks if necessary

````python
def check_protected(amount: int | float) -> print: return print(f'{amount:{"*"}>10,}')

amount_one = 399.87
amount_two = 17_934.3
amount_three = 1_140_999.34

check_protected(amount_one)
check_protected(amount_two)
check_protected(amount_three)
````

8. 3 ***(Encrypt a Phrase)*** In Exercise 5.11, you wrote a function to calculate an encryption key based on a given sentence. Use this function to generate an encryption key to help encrypt a phrase. Write a function to encrypt the phrase using the following algorithm:\
&nbsp;&nbsp;&nbsp; To encrypt a phrase, you need to tokenize it using the string method *split*. After this, you can encrypt each word individually. Search for the index of each letter of the word in the alphabet. Select the letter in the encryption key with the same index. Replace the original letter in the word with the encrypted letter. Replace the original word in the tokenized sentence with the encrypted word. If the word is only one letter long, it will not be encrypted. When all the individual words are encrypted, they are combined into a new sentence making sure that blanks between words remain blanks. Enable the user to enter a phrase and subsequently display the encrypted phrase.


````python


def encryption_sentence(sentence: str) -> str:
   '''It takes a sentence, splits it into words, encrypts each word, and then joins the encrypted words
   back into a sentence
   
   Parameters
   ----------
   sentence : str
      str
   
   Returns
   -------
      The encrypted sentence.
   
   '''
   def encryption_key(string):
    '''Creates an encryption key by isolating the unique letters of the string,
       if it does not have all the letters of the alphabet are added'''

    #Initialization Phase
    lowercase_alphabet, uppercase_alphabet = map(chr,range(97,123)) ,map(chr,range(65,91))
    code_key = list() 

    #Processing Phase
    list(map(lambda letter: code_key.append(letter) if code_key.count(letter.lower()) < 1 and letter.upper() not in code_key else None,
            filter(lambda letter : letter.isalpha()  ,string ))) 
    #The filter gets only the letters and the mapping adds them only once
   
    list(map(lambda letter_lower, letter_upper: code_key.append(letter_lower) if letter_lower not in code_key and letter_upper not in code_key else None, lowercase_alphabet,uppercase_alphabet)) 
    #Adds the missing letters of the alphabet
   
    #Termination Phase
    return ' '.join(code_key)
   
   new_sentence = list()

   for word in sentence.split():
      
      alphabet = list(map(chr, range(97, 123)))
      
      word_encrypted = encryption_key(word).replace(' ', '').lower()
      
      new_word = str()
    
      if len(word) > 1:
         for letter in word.lower(): 
            new_word += word_encrypted[alphabet.index(letter)] if letter.isalpha() else letter
      else:
            new_word += word
      
      new_sentence.append(new_word)
   

   return ' '.join(new_sentence).capitalize()

sentence = "How funny it'll seem to come out among the people that walk with their heads downwards! The antipathies, I think"

print("The sentence",sentence, sep ='\n')
print(f"\nThe encrypted sentence")
print(encryption_sentence(sentence))
      
````

8. 10 ***(Project: Simple Sentiment Analysis)*** Search online for lists of positive sentiment words and negative sentiment words. Create a script that inputs text, then determines whether that text is positive or negative, based on the total number of positive words and the total number of negative words. Test your script by searching for Twitter tweets on a topic of your choosing, then entering the text for several tweets. In the data science case study chapters, weâ€™ll take a deeper look at sentiment analysis

````python
from nltk.corpus import twitter_samples

def checks_sentence_sentiment(sentence: str):
    '''The function takes a sentence and returns a string that says whether the sentence is
    positive, negative, or neutral
    
    Parameters
    ----------
    sentence
        The sentence you want to check the sentiment of.
    
    Returns
    -------
        a string that tells the user whether the sentence is positive, negative, or neutral.
    
    '''
    from nltk.corpus import stopwords
    from pandas import read_csv
    from re import sub
    
    
    positive_words = read_csv("PositiveWords.txt") #dataset with positive words.

    negative_words = read_csv("negative_words.csv") # dataset with negative words.
    
    sentence = sub(r"([@!#_,]*)(\d*)", "", sentence).split() # deletes any character that is not a word or emoticon character.
    stopwords = stopwords.words("english") 
    
    post_emoticon =  [':-)', ':)', ';)', ':o)', ':]', ':3', ':c)', ':>', '=]', '8)', '=)', ':}',
                      ':^)', ':-D', ':D', '8-D', '8D', 'x-D', 'xD', 'X-D', 'XD', '=-D', '=D',
                      '=-3', '=3', ':-))', ":'-)", ":')"]
    
    neg_emoticon = [':L', ':-/', '>:/', ':S', '>:[', ':@', ':-(', ':[', ':-||', '=L', ':<',
                     ':-[', ':-<', '=\\', '=/', '>:(', ':(', '>.<', ":'-(", ":'(", ':\\', ':-c',
                     ':c', ':{', '>:\\', ';(',]
    
    count_positive = int()
    count_negative = int()
    
    for word in filter(lambda word: word.lower() not in stopwords, sentence):
        
        if word.lower() in positive_words.values or word in post_emoticon :
            count_positive += 1
        elif word.capitalize() in negative_words.values or word in neg_emoticon:
            count_negative += 1
            
    if count_positive > count_negative:
        return "This sentence is Positive"
    elif count_negative > count_positive:
        return "This sentence is Negative"
    else:
        return "This sentence is Neutral"
    
    
tweets_post_one = twitter_samples.strings("positive_tweets.json")[0]
tweets_post_two = twitter_samples.strings("positive_tweets.json")[1]
tweets_post_three = twitter_samples.strings("positive_tweets.json")[2]

tweets_neg_one = twitter_samples.strings("negative_tweets.json")[0]
tweets_neg_two = twitter_samples.strings("negative_tweets.json")[1]
tweets_neg_three = twitter_samples.strings("negative_tweets.json")[2]


print("1) Positive tweets\n", tweets_post_one)
print(checks_sentence_sentiment(tweets_post_one), end="\n\n")
print("2) Positive tweets\n", tweets_post_two)
print(checks_sentence_sentiment(tweets_post_two), end="\n\n")
print("3) Positive tweets\n", tweets_post_three)
print(checks_sentence_sentiment(tweets_post_three), end="\n\n")

print()
print("1) Negative tweets\n", tweets_neg_one)
print(checks_sentence_sentiment(tweets_neg_one), end="\n\n")
print("2) Negative tweets\n", tweets_neg_two)
print(checks_sentence_sentiment(tweets_neg_two), end="\n\n")
print("3) Negative tweets\n", tweets_neg_three)
print(checks_sentence_sentiment(tweets_neg_three), end="\n\n")
````

8. 11 ***(Project: Evaluate Word Problems)*** Write a script that enables the user to enter mathematical word problems like *"two times three"* and  *"seven minus five"*, then use string processing to break apart the string into the numbers and the operation and return the result. So *"two times three"* would return 6 and *"seven minus five"* would return 2. To keep things simple, assume the user enters only the words for the numbers 0 through 9 and only the operations *'plus'*, *'minus'*, *'times'*,  and *'divided by'*.

````python

def evaluate_word_problem(problem: str) :
    '''It takes a string, splits it into a list, then uses a dictionary to convert the words into numbers
    and operators, then evaluates the expression
    
    Parameters
    ----------
    problem : str
        str
    
    '''

    numbers = {k: str(v) for v,k in enumerate(("Zero", "One", "Two", "Three",
                                                "Fourth", "Five", "Six", 
                                                "Seven", "Eight", "Nine"))}

    operators = {"Plus": "+", "Minus": "-", "Times": "*", "Divided By": "/"}
    
    operation = problem.title().split()
    
    if len(operation) == 3: 
    
        operation =  numbers.get(operation[0]) ,\
                    operators.get(operation[1]) , \
                    numbers.get(operation[2])
        
    else:
         operation =  numbers.get(operation[0]) , \
                      operators.get(' '.join((operation[1], operation[2]))) , \
                      numbers.get(operation[3])
    
    
    return f"Result = {eval(''.join(operation))}" if all(operation) else "incorrect operation"
        
    
problem_one = " two times three  "
problem_two= "eight divided By fourth "
problem_three = "seven minus  five"
problem_fourth = "five minus o2ne"
  
  
print("1)", problem_one, end="\n")
print(evaluate_word_problem(problem_one))
print("2)", problem_two, end="\n")
print(evaluate_word_problem(problem_two))
print("3)", problem_three, end="\n")
print(evaluate_word_problem(problem_three))
print("4)", problem_fourth, end="\n")
print(evaluate_word_problem(problem_fourth))
````

8. 12 ***(Project: Scrambled Text)*** Use string-processing capabilities to keep the first and last letter of a word and scramble the remaining letters in between the first and last. Search online for "University of Cambridge scrambled text" for an intriguing paper on the readability of texts consisting of such scrambled words. Investigate the *random* module's *shuffle* function to help you implement this exercise's solution.

````python
def scrambled_text(text: str):
    '''It scrambles the middle letters of each word
    
    Parameters
    ----------
    text : string
    
    Returns
    -------
        The scrambled text.
    
    '''
    from random import sample
    new_text = str()
    
    for words in text.split():
        
        if len(words) > 2:
            
            if words[-1].isalnum():
        
                new_text += " " + words[0] + ''.join(sample(words[1:-1], len(words) - 2)) + words[-1]
                
            else:
  
                new_text += " " + words[0] + ''.join(sample(words[1:-2], len(words) - 3)) + words[-2:]
            
        else:
            new_text +=  " " + words
    
    return new_text.strip()
            

text = """According to a researcher (sic) at Cambridge University, it doesn't matter in what order the letters in a word are,
        the only important thing is that the first and last letter be at the right place. The rest can be a total mess and you can still read it without problem.
        This is because the human mind does not read every letter by itself but the word as a whole."""
        
print(scrambled_text(text))
````

8. 18 ***(Using Regular Expressions for Password Validation)*** When choosing a password, a user is often confronted with a set of rules they have to comply with in order to set a valid password. Write a script using the necessary regular expressions to check if a password is valid. For a chosen password to be valid, it should contain at least one lowercase and one uppercase letter and at least one number. The use of whites paces in a password is not allowed. Provide the necessary feedback to the user for their chosen password.

````python

def check_password(password: str) -> str:
    """Check if the password is valid, if not prints out the reason why it's not valid"""
    from re import fullmatch, search 

    pattern = r"(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?!.*\s)(\S+)"

    if search(pattern, password):
        return "VALID"

    else:
        match True:
            case True if fullmatch("[^A-Za-z0-9]+", password):
                print("The password is valid if contain:")
                print("\t1) at least one uppercase letter".expandtabs(33))
                print("\t2) at least one lowercase letter".expandtabs(33))
                print("\t3) at least one number".expandtabs(33))
                print("\t4) not contains whitespaces".expandtabs(33))
                return ""

            case True if not search("[A-Z]", password):
                print("The password should contain at least one uppercase letter")
                return ""
            case True if not search("[a-z]", password):
                print("The password should contain at least one lowercase letter")
                return ""
            case True if not search("\d", password):
                print("The password should contain at least one number")
                return ""
            case True if search("\s", password):
                print("Whitespaces in the password is not allowed")
                return ""

passwords = "numbEr3", "a@M1?", "number2", "Python 3", "?@"

for password in (passwords):
    print(f"Password: {password}",  end="\n")
    print(check_password(password))
````

````python

def munging_date(text: str) -> str:
    '''Transform dates in the text to other formatting 
    
    Parameters
    ----------
    Text: str
        Text with dates of type dd-mm-yyyy, mm/dd/yyyy or MYD
    Returns
    -------
        The text with its dates formatted to a different format 

    '''
    import re

    def change_format(date: str):
        '''Determines the type of format in which the date is represented and changes it to another format.
            
        Parameters
        ----------
        date:
            The date 
        
        Returns
        --------
        The date in a different format
        '''
        
        from random import randint

        def formats(typ: int):
            '''Change the representation of the date to a different format 
            
             Parameters
            ----------
            typ:
                1 = dd-mm-yyyy.
                2 = mm/dd/yyyy.
                3 = MDY.
            Returns
            -------
                The representation of the date.
            '''
            
            match typ:
                case 1:
                    return '-'.join(f"{dat:0>2}" for dat in (day, month, year))
                case 2:
                    return '/'.join(f"{dat:0>2}" for dat in (month, day, year))
                case 3:
                    return months[int(month) - 1] + ",".join(f" {dat:0>2}" for dat in (day, year))

        months = ("January", "February", "March", "April", "May",
                  "June", "July", "August", "September", "October",
                  "November", "December")

        if re.match(pattern_format1, date): # Checks if it matches the format type MDY, (e.g July 11, 1987).

            month, day, year = re.fullmatch(r"([A-Z][a-z]+)\s+(\d{2})\,\s+(\d{4})", date).groups() # splits the string into month, day, year
            month = months.index(month) + 1 # Change the month into its numerical equivalent, (e.g, November to 8) 
            
            #If the number is even, the date is transformed to dd-mm-yyyy format(e.g 12-10-2022), otherwise it is transformed to mm/dd/yyyy
            return formats(typ=1) if (randint(0, 100) % 2) == 0 else formats(typ=2)  

        else:
            month, day, year = re.fullmatch(r"(\d{1,2})\S(\d{1,2})\S(\d{4})", date).groups() # splits the string into month, day, year

            if "-" in date: # Checks if date is not in the format dd-mm-yyyy
                day, month, = month, day # if so, it changes them 
                
                #If the number is odd, the date is transformed to MDY format(e.g May 12, 2011), otherwise it is transformed to mm/dd/yyyy
                return formats(typ=3) if (randint(0, 100) % 2) != 0 else formats(typ=2)

            else:
                #If the number is odd, the date is transformed to MDY format(e.g January 12, 2023), otherwise it is transformed to dd-mm-yyyy
                return formats(typ=3) if (randint(0, 100) % 2) != 0 else formats(typ=1)

    pattern_format1 = r"([A-Za-z]{3,9}\s+\d{1,2},\s+\d{4})" # MDY format pattern
    pattern_format2 = r"(\d{1,2}/\d{1,2}/\d{4})" # mm/dd/yyyy format pattern
    pattern_format3 = r"(\d{1,2}-\d{1,2}-\d{4})" # dd-mm-yyyy format pattern

    old_text = text

    for pattern in (pattern_format1, pattern_format2,
                    pattern_format3):
        
        # if the pattern matches at least once in the text
        if re.search(pattern, old_text): # the operation is performed

            for old_date in re.finditer(pattern, old_text): 

                new_date = change_format(old_date.group())

                if text.count(old_date.group()) >= 2: 

                    index = text.rfind(old_date.group()) # if there are two or more occurrences in the text,
                    text = text[:index] + text[index:].replace(old_date.group(), new_date) # the last one is changed first.
                    
                else:
                    text = text.replace(old_date.group(), new_date)

    return text


test_one = "25-04-1955"
test_two = "3/25/1955 ; /01/24/2021"
test_three = "Date:April 25, 1955 :) August 31, 1955. Dtes: November 25, 1445 or 01/24/2021= ~~Â£ and25-04-1955"

test_fourth = "In 05/02/1946, a model for computer architecture was introduced and became known as Von Neumann architecture. Since May 12, 1950, "\
    "the von Neumann model provided uniformity in subsequent computer designs. "\
    "The von Neumann architecture was considered innovative as it introduced an idea of allowing machine instructions and data to share memory space. "\
    "The von Neumann model is composed of three major parts, the arithmetic logic unit (ALU), the memory, and the instruction processing unit (IPU)." \
    "In von Neumann machine design, the IPU passes addresses to memory, and memory, in turn, is routed either back to the IPU if an instruction is being fetched or to the ALU if data is being fetched."

for date in (test_one, test_two, test_three, test_fourth):

    print(f"Before \n{date}\n")
    print(f"After \n{munging_date(date)}")
    print("_" * len(date), end="\n") if date != test_fourth else None
````
