Then,call the function three times starting with the numbers 0, 1, and 1 and restarting the function each time with the resulting numbers of the previous iteration.

```python
def Fibonnaci(a,b,c):
    '''
        Receives three consecutive numbers of Fibonnaci's series
        and returns the next three numbers
    '''
    
    for x in range(3):
        
        next = b + c #Next number of the serie
        b, c = c, next #Anteultimate and last issue of the serie
        
        yield next

print(list(Fibonnaci(0,1,1)))
print(tuple(Fibonnaci(2,3,5)))
```

5. 8 ***(Insertion Sort Algorithm)*** In computer science, there are several algorithms available to sort the elements of a list such as quicksort, merge, heapsort, and insertion sort.Although insertion sort is not the most efficient sorting algorithm when dealing with large lists, it is very efficient for small data sets and easy to implement. The algorithm splits a list into a sorted and unsorted part. Each value from the unsorted part is sequentially compared to each element in the sorted part and placed in the correct position. The process to sorting the list is:
- *a)* Get a list of unsorted numbers.
- *b)* Compare the first two elements.
- *c)* If they are in descending order, switch them. This is now your sorted sub-list.
- *d)* Take the third element of the list Compare, and if necessary, swap it with the second element. Now compare the new second element and swap, if necessary,with the first element
- *e)* Take the fourth element of the list.Compare and swap with third element,if necessary.
- *f)* Keep repeating Steps *(a)* to *(e)* until every elemnt of the list is in its correct place.

When this process completes, the list of numbers is sorted in ascending order.Write a function *insertion sort* implementing the described algorithm. Use a list of 10 unique and randomly picked numbers between 0 and 100 to test this function. Display the unsorted and the sorted list to evaluate the validity of your function

```python
import random

def insertionSort(array: list) -> list:
    '''
        Sort list  of comparable elements into nondecrasing order
    '''
    
    for i in range (1, len(array)):

         while (i > 0) and (array[i-1] > array[i]) : 
            #Move elements of array[0..i-1],that are greater, 
            #to one position ahead of their current position
            
            array [i], array[i-1]= array[i-1],array[i]
            i -= 1
   
    return array      


A =[random.randrange(101) for x in range(10)] 
print(f'List ---> {A}')
print(f'This is Insertionsort: {insertionSort(A)}')
```

Write a function is_palindrome that takes a string and returns True if it’s a palindrome and False otherwise. Your function should ignore case sensitivity (that is, 'a' and 'A' are the same) ,spaces and punctuation.

```python
def is_palindrome(string,key= bool):
    '''Returns True if the string is palindrome otherwise False.
       
       If key is True,returns the cleaned word as a list of its letters
        '''
    
    string_cleaner = list( map(lambda s: s.upper(), filter(lambda w: w.isalpha(), string) ) ) #The filter gets only the letters and the mapping converts them to upper case letters

    print(string_cleaner) if key is True else None

    return string_cleaner == list(reversed(string_cleaner))

string_one= '.RA3d,!ar3.' # 'Radar' --> Palindrome
string_two= 'Si+g°mA' # 'Sigma'-->Non-Palindrome
string_three = 'Doc, note, I dissent. A fast never prevents a fatness. I diet on cod.' #--> Palindrome

print(f'This string "{string_one}" is a palindrome: {is_palindrome(string_one,key= True)}')

print(f'This string "{string_two}" is a palindrome: {is_palindrome(string_two)}')

print(f'This string "{string_three}" is a palindrome: {is_palindrome(string_three)}')
```

5. 25 **(Card Playing: Evaluating Poker Hands)** Modify Exercise 5.24 to deal a five-card poker hand as list of five card tuples. Then create functions (i.e.,  *is_pair*, *is_two_pair*, *is_three_of_a_kind*, ...) that determine whether the hand they receive as an argument contains groups of cards, such as:
- a) one pair 
- b) two pairs 
- c) three of a kind (e.g., three jacks)
- d) a straight (i.e., five cards of consecutive face values)
- e) a flush (i.e., all five cards of the same suit)
- f) a full house (i.e, two cards of one face value and three cards of another)
- g) four of a king (e.g., four aces)
- h) straight flush (i.e., a straight with all five cards of the same suit)
- i) ... and others.

See https://en.wikipedia.org/wiki/List_of_poker_hands for poker-hand types and how they rank with respect to one another. For example, three of a kind beats two pairs.

```python
def poker_hand(hand):
    """ Depending on the combination of cards, returns the poker hand it represents. """
    
    straight = [ 0,0,'Deuce', 'Three', 'Four','Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace' ]
    
    flush =  list( map( lambda i: i.endswith( hand[1].split()[2] ), hand ) ) # Checks if they are all of the same suit by comparing the last word of each card.
    straight_test =  list ( map( lambda i: straight.index( i.split()[0] ), hand) ) # Obtains the indexes of each card in the straight variable
    same_value = list( map( lambda j: straight_test.count(j), straight_test) ) # Gets the number of times a card is repeated in the straight_test variable
    
    if  all(flush)  and (min(straight_test) == 10 ): # If they are all of the same suit and the smallest card is ten, it is a royal flush.
        
        return ( f'{ "Royal flush" :>40}\n \n{ ", ".join(hand) }' )
   
    elif all(flush) and ( max(straight_test) - min(straight_test) == 4 ):  # If they are all of the same suit 
                                                                 # and the difference between the highest card and the lowest card is 4 (straight),
        return ( f'{ "Straight flush" :>40}\n \n{ ", ".join(hand) }' )  # it is a straight flush.
        
    elif ( 4 in same_value):  # If a card is repeated four times it means there is a four of a king.
    
        return ( f'{ "Four of a kind" :>40}\n \n{ ", ".join(hand) }' )
    
    elif ( 3 in same_value) and ( 2 in same_value): # Two cards of one face value and three cards of another.
        
        return ( f'{ "Full House" :>40}\n \n{ ", ".join(hand) }' )
    
    elif all(flush): 
        
        return ( f'{ "Flush" :>40}\n  \n{ ", ".join(hand) }' )
    
    elif ( max(straight_test) - min(straight_test) == 4 ) and ( same_value.count(1) == 5 ) : # Checks that each card is unique, 
                                                                                             # in case one is repeated with the same value as another.
        return( f'{ "Straight" :>40}\n \n{ ", ".join(hand) }' )
    
    elif ( 3 in same_value): # It is a third of a king if there are three cards of the same rank.
        
        return ( f'{ "Three of a kind" :>40}\n \n{ ", ".join(hand) }' )
   
    elif (same_value.count(2) == 4 ): # If there are two cards ,and each card is repeated twice, 
                                    # the total of 2 in the same_value variable will have to be four (one for each card).
        return ( f'{ "Two pair" :>40}\n \n{ ", " .join(hand) }' )
        
    elif ( 2 in same_value): # One pair 
        
        return  ( f'{ "One pair" :>40}\n \n{ ", ".join(hand) }' )
        
    else: # High card 
        
        return ( f'{ "High card" :>40}\n \n{ ", ".join(hand) }' )
```

5. 33 **(Intro to Data Science: Rolling Two Dice)** In this exercise, you’ll modify Chapter 4’s script that simulates the dice game craps by using the techniques you learned in Section 5.17.2. The script should receive a command-line argument indicating the number of games of craps to execute and use two lists to track the total numbers of games won and lost on the first roll, second roll, third roll, etc.
- Display a horizontal bar plot indicating how many games are won and how
many are lost on the first roll, second roll, third roll, etc. Since the game could
continue indefinitely, you might track wins and losses through the first dozen
rolls (of a pair of dice), then maintain two counters that keep track of wins and
losses after 12 rolls—no matter how long the game gets. Create separate bars
for wins and losses

```python
import numpy as np
import matplotlib.pyplot as plt

 #Initialization Phase
def game_craps() :
    
    """Simulating the dice game Craps."""
        
    def roll_dice():
        """Roll two dice and return their face values as a tuple."""
        import random
        
        die1 = random.randrange(1, 7)
        die2 = random.randrange(1, 7)
        return (die1, die2)  #
    
    die_values = roll_dice()  # first roll
    
    # determine game status and point, based on first roll
    sum_of_dice = sum(die_values)

    if sum_of_dice in (7, 11):  # win
        game_status = 'WON'
        return 1
    elif sum_of_dice in (2, 3, 12):  # lose
        game_status = 'LOST'
        return 0
    else:  # remember point
        game_status = 'CONTINUE'
        my_point = sum_of_dice
        # print('Point is', my_point)

    # continue rolling until player wins or loses
    while game_status == 'CONTINUE':
        die_values = roll_dice()
    
        sum_of_dice = sum(die_values)

        if sum_of_dice == my_point:  # win by making point
            game_status = 'WON'
            # print('Player wins')
            return 1
        
        elif sum_of_dice == 7:  # lose by rolling 7
            game_status = 'LOST'
            # print('Player loses')
            return 0
        
#Processing Phase
number_of_games = np.random.randint(100,1000,12) # Integers representing the number of times game_craps must be played in each round. 
rounds =  map(lambda x: [game_craps() for i in range(x) ], number_of_games )  #Play game_craps N-times per round
frequencies = list( map(lambda x: np.unique(x,return_counts=True)[1], rounds ) )# List with the number of games won and lost of each round
won = list(map( lambda x: x[1], frequencies)) #Games won in each round
lost = list(map( lambda x: x[0], frequencies)) #Games lost in each round

#Termination Phase
x = np.arange(len(number_of_games)) #Number of rounds ,to be used in x-axis and y-axis
width = 0.40

plt.style.use("fivethirtyeight")
plt.tight_layout()
plt.grid()

games_won = plt.barh(x  - width/2, won, width, color ="#0000ff", label="Games Won")
games_lost = plt.barh(x + width/2, lost, width, color="#ff0000", label="Games lost")

plt.title(f'games won and lost\n in {len(x)} rounds of craps game'.title()) 
plt.ylabel('Games Played') 
plt.yticks(ticks= x, labels= number_of_games)

plt.bar_label(games_won, padding=3, size= 9)
plt.bar_label(games_lost, padding=3, size= 9)

plt.ylim(-1,len(number_of_games) )
plt.xlim(0,max((max(x) for x in frequencies)) + min((min(x) for x in frequencies)))

plt.legend()
plt.show()
```
